```python
import numpy as np
from scipy.stats import beta 
from matplotlib import pyplot as plt
```

```python
data = np.load('D:\TU Darmstadt\SS23\SML\Homework 2\SML2023-Homework2\data\\binomial_data.npy')
```

```python
def beta_ploter(a,b,N):
    x=np.linspace(0,1,1000)
    plt.plot(x, beta.pdf(x,a,b), label = "N=%d"%N)
    
def param_update(data):
    #Starting point
    a=1
    b=1
    
    for point in data:
        a+=point
        b+=(10-point)
    return a, b
        
N_values = np.array([1, 2, 5, 10, 100])

def main():
    for N in N_values:
        a, b = param_update(data[:N])
        beta_ploter(a,b,N)
    plt.legend(loc="upper right")


main()
y = np.linspace(0, 37, 1000)
plt.axvline(x = 0.165099999999999, color = 'b', label = 'MLE parameter for N=1000')
plt.legend(loc="upper right")
```

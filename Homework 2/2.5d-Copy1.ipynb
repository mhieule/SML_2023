```python
import numpy as np
from matplotlib import pyplot as plt
import math
data = np.load('D:\TU Darmstadt\SS23\SML\Homework 2\SML2023-Homework2\data\data.npy')
label = np.load('D:\TU Darmstadt\SS23\SML\Homework 2\SML2023-Homework2\data\labels.npy')
def predict(w, b, x):
    return -1 if (w.dot(x)+b) < 0 else 1
def label_colormap(label):
    result  = [0]*label.size
    for i in range(label.size):
        result[i] = 0 if label[i]==-1 else 1
    return result        
def feature_transform(data, label):
    x, y = np.copy(data[:,0]), np.copy(data[:,1])
    data[:,0] = np.multiply(x, np.cos(y))
    data[:,1] = np.multiply(x, np.sin(y))    
feature_transform(data, label)
colormap = np.array(["r", "k"])
titlemap=np.array(["label=-1","label=1"])
plt.scatter(data[:,0], data[:,1], c=colormap[label_colormap(label)])
def perceptron(data, label, n=1000):
    weight, bias = np.array([0,0]), 0
    for iter in range(n):
        i = 4
        if(predict(weight, bias, data[i]) != label[i]):
            if(label[i] == 1):
                weight = weight + data[i]
                bias += 1
            elif(label[i]==-1):
                weight = weight - data[i]
                bias -= 1
    return weight, bias
weights, bias = perceptron(data, label, 1000)
w_0, w_1 = weights[0], weights[1]
slope, intercept = -w_0/w_1, -bias/w_1 
xx = np.linspace(-10, 10, 100000)
yy = xx * slope + intercept
plt.plot(xx, yy, "k-")
plt.scatter(data[4][0], data[4][1], color="blue", label = "5-th point")
plt.legend(loc=0)
```

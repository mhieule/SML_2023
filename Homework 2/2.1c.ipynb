# Import libraries and load data

```python
import numpy as np
import matplotlib.pyplot as plt
import math


data = np.load('D:\TU Darmstadt\SS23\SML\Homework 2\SML2023-Homework2\data\\binomial_data.npy')
print((data))
```

## Function to count the occurences of 1-10 in a given list

```python
#Function to count the occurences of how many times each outcome of X is recorded (from 1-10), normalized.
def count_occurences(arr,N):
    counts = [0]*11
    for num in arr:
        counts[num]+=1 
    counts = np.divide(counts,N)
    return counts

#Function to calculate the maximum likelihood parameter
def ml_param(dataset, N, n=10):
    data_sum = sum(np.divide(dataset,n))
    return data_sum/N

#Maximum likelihood parameter for N=10,100,1000 data points
ml_param_10 = ml_param(data[:10], 10, 10)
ml_param_100 = ml_param(data[:100], 100, 10)
ml_param_1000 = ml_param(data[:1000], 1000, 10)
print((ml_param_10))
print(ml_param_100)
print(ml_param_1000)

#Function to pmf for N=10,100,1000
def pmf(theta, x, n=10):
    return math.comb(n,x)*(theta**x)*((1-theta)**(n-x))

#Outcomes
out = np.arange(0,11)

pmf_10 = [pmf(ml_param_10, elem_out, 10) for elem_out in out]
pmf_100 = [pmf(ml_param_100, elem_out, 10) for elem_out in out]
pmf_1000 = [pmf(ml_param_1000, elem_out, 10) for elem_out in out]


#Plot the frequency of each outcome
plt.plot(count_occurences(data[:10],10), label = "Relative frequency of outcomes")
plt.plot(pmf_10, label = "PMF with ML parameter")
plt.legend(loc=0)
plt.title(label="N=10")
```

```python
plt.plot(count_occurences(data[:100],100), label = "Relative frequency of outcomes")
plt.plot(pmf_100, label = "PMF with ML parameter")
plt.legend(loc=0)
plt.title(label="N=100")
```

```python
plt.plot(count_occurences(data[:1000],1000), label = "Relative frequency of outcomes")
print(count_occurences(data[:1000],1000))
plt.plot(pmf_1000, label = "PMF with ML parameter")
plt.legend(loc=0)
plt.title(label="N=1000")
```

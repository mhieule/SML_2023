```python
import numpy as np
from matplotlib import pyplot as plt
import math

data = np.load('D:\TU Darmstadt\SS23\SML\Homework 2\SML2023-Homework2\data\data.npy')
label = np.load('D:\TU Darmstadt\SS23\SML\Homework 2\SML2023-Homework2\data\labels.npy')

np.random.seed(1)

def predict(w, b, x):
    return -1 if (w.dot(x)+b) < 0 else 1

def perceptron(data, label, n=1):
    weight = np.array([0,0])
    bias = 0
    for iter in range(n):
        i = 4
        if(predict(weight, bias, data[i]) != label[i]):
            if(label[i] == 1):
                weight = weight + data[i]
                bias += 1
            elif(label[i]==-1):
                weight = weight - data[i]
                bias -= 1
    return weight, bias

weights, bias = perceptron(data, label, 1)
w_0 = weights[0]
w_1 = weights[1]
slope = -w_0/w_1
intercept = -bias/w_1

def label_colormap(label):
    result  = [0]*label.size
    for i in range(label.size):
        if(label[i]==-1):
            result[i]=0
        else: result[i]=1     
    return result
        
colormap = np.array(["r", "k"])
plt.scatter(data[:,0], data[:,1], c=colormap[label_colormap(label)])
xx = np.linspace(0, 13, 1000)
yy = xx * slope + intercept
plt.plot(xx, yy, "k-")
plt.scatter(9.60025098, 3.90141117, color="yellow")
plt.quiver([0],[0],[9.60025098],[3.90141117], color='b', units='xy', scale=1)

plt.xlim(0, 12)
plt.ylim(0, 5)

print(f"weights{weights}")
print(f"bias{bias}")
print(f"slope{slope}")
print(f"intercept{intercept}")
print(data)
```
